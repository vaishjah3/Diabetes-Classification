```python
import numpy as np
import torch
import torch.nn as nn
import pandas as pd
from sklearn.preprocessing import StandardScaler
from torch.utils.data import Dataset
```

```python
data=pd.read_csv('diabetes.csv')
data
```

```python
x=data.iloc[:,:-1]
x
```

```python
y_string=data.iloc[:,-1]
y_int=y_string.astype('category').cat.codes
y_int
```

```python
y=np.array(y_int, dtype='float64')
y
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x=torch.tensor(x)
y=torch.tensor(y).unsqueeze(1)
```

```python
print(x.shape)
print(y.shape)
```

```python
class Dataset(Dataset):
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __getitem__(self, index):
    #get one item from dataset
    return self.x[index], self.y[index]
  def __len__(self):
    return len(self.x)
```

```python
dataset=Dataset(x,y)
```

```python
len(dataset)
```

```python
train_loader=torch.utils.data.DataLoader(dataset=dataset,
                                         batch_size=32,
                                         shuffle=True)
```

```python
print(train_loader)
```

```python
print(len(train_loader))
```

```python
for (x,y) in train_loader:
  print(x.shape)
  print(y.shape)
  break
```

```python
class Model(nn.Module):
  def __init__(self, input_features, output_feature):
    super(Model, self).__init__()
    self.fc1=nn.Linear(input_features, 5)
    self.fc2=nn.Linear(5,4)
    self.fc3=nn.Linear(4,3)
    self.fc4=nn.Linear(3, output_feature)
    self.sigmoid=nn.Sigmoid()
    self.tanh=nn.Tanh()

  def forward(self, x):
    out=self.fc1(x)
    out=self.tanh(out)
    out=self.fc2(out)
    out=self.tanh(out)
    out=self.fc3(out)
    out=self.tanh(out)
    out=self.fc4(out)
    out=self.sigmoid(out)
    return out
```

```python
net=Model(7,1)
criterion=torch.nn.BCELoss(size_average=True)
```

```python
optim=torch.optim.SGD(net.parameters(), lr=0.1, momentum=0.9)
```

```python
epochs=10
for epoch in range(epochs):
  for inputs, labels in train_loader:
    inputs=inputs.float()
    labels=labels.float()
    output=net.forward(inputs)
    loss=criterion(output, labels)
    optim.zero_grad()
    loss.backward()
    optim.step()

  output=(output>0.5).float()
  accuracy=(output==labels).float().mean()
  print(loss)
  print(accuracy)
```
